```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("super_store_sales.csv",encoding='latin1')
df.head(10)
```

```python
df.info()
```

```python
df['OrderDate']=pd.to_datetime(df['OrderDate'])
df['ShipDate']=pd.to_datetime(df['ShipDate'])
```

```python
df.info()
```

```python
# Extract Year
df['Year'] = df['OrderDate'].dt.year

# Extract Month Number
df['Month'] = df['OrderDate'].dt.month

# Extract Month Name (e.g., Jan, Feb)
df['Month_Name'] = df['OrderDate'].dt.month_name()

# Extract Quarter (Q1, Q2, Q3, Q4)
df['Quarter'] = df['OrderDate'].dt.quarter

# Extract Day of Week Number (0=Monday, 6=Sunday)
df['Weekday'] = df['OrderDate'].dt.weekday

# Extract Day Name (Monday, Tuesday…)
df['Weekday_Name'] = df['OrderDate'].dt.day_name()

# Calculate Shipping Duration (days between order and ship date)
df['Ship_Duration'] = (df['ShipDate'] - df['OrderDate']).dt.days
```

```python
df.info()
```

```python
weekday_sales=df.groupby('Weekday_Name')['Sales'].sum()
weekday_sales=weekday_sales.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
weekday_sales
plt.figure(figsize=(8,5))
sns.barplot(x=weekday_sales.index,y=weekday_sales.values)
plt.title("total sale by weekday")
plt.ylabel("total sales")
plt.show()
```

```python
order = ['January','February','March','April','May','June',
         'July','August','September','October','November','December']
monthly_sales=df.groupby('Month_Name')['Sales'].sum().reindex(order)
monthly_sales
sns.lineplot(x=monthly_sales.index,y=monthly_sales.values,marker="o")
plt.title("montly sales trends")
plt.xlabel("month")
plt.ylabel("total sales")
plt.xticks(rotation=45)
plt.show()
```

```python
region_sales=df.groupby('Region')['Sales'].sum()
plt.figure(figsize=(8,5))
sns.barplot(x=region_sales.index,y=region_sales.values)
plt.title("sale by region")
plt.xlabel("region")
plt.ylabel("total sale")
plt.show()
```

```python
top_product=df.groupby('Product Name')['Sales'].sum().nlargest(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_product.values,y=top_product.index)
plt.title("Top 10 Products by Sales")
plt.xlabel("Total Sales")
plt.show()
```

```python
category_profit=df.groupby('Category')["Profit"].sum()
plt.figure(figsize=(7,5))
sns.barplot(x=category_profit.index, y=category_profit.values)
plt.title("Profit by Category")
plt.ylabel("Total Profit")
plt.show()
```

```python
##discount vs profit
plt.figure(figsize=(8,6))

sns.scatterplot(x="Discount",y="Profit",data=df,alpha=0.6)
plt.title("Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.show()
```

```python
year_sales = df.groupby('Year')['Sales'].sum()

plt.figure(figsize=(8,5))
sns.barplot(x=year_sales.index, y=year_sales.values)
plt.title("Yearly Sales Trend")
plt.ylabel("Total Sales")
plt.show()
```

```python
quarter_sales = df.groupby('Quarter')['Sales'].sum()

plt.figure(figsize=(7,5))
sns.barplot(x=quarter_sales.index, y=quarter_sales.values)
plt.title("Quarterly Sales Trend")
plt.xlabel("Quarter")
plt.ylabel("Total Sales")
plt.show()
```

```python
weekday_sales = df.groupby('Weekday_Name')['Sales'].sum()

# Reorder weekdays
weekday_sale= weekday_sales.reindex(
    ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
)

plt.figure(figsize=(9,5))
sns.barplot(x=weekday_sale.index, y=weekday_sale.values)
plt.title("Sales by Weekday")
plt.ylabel("Total Sales")
plt.show()
```

```python
shipping_duration=df.groupby("Region")['Ship_Duration'].mean()

plt.figure(figsize=(8,5))
sns.barplot(x=shipping_duration.index, y=shipping_duration.values)
plt.title("Average Shipping Duration by Region")
plt.ylabel("Avg Shipping Days")
plt.show()
```
